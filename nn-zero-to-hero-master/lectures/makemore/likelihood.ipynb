```python
import numpy as np
import matplotlib.pyplot as plt

# 定义 x 的范围，从非常小的正数开始到 1
x = np.linspace(0.01, 1, 100)  # 从 0.01 开始以避免 log(0) 的问题
y = np.log(x)  # 计算自然对数

# 绘制图形
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = ln(x)', color='blue')
plt.title('Natural Logarithm Function')
plt.xlabel('x')
plt.ylabel('ln(x)')
plt.xlim(0, 1)
plt.ylim(-5, 0)  # 根据自然对数的范围调整 y 轴
plt.axhline(0, color='black', lw=0.5, ls='--')  # 添加 y=0 的水平线
plt.axvline(0, color='black', lw=0.5, ls='--')  # 添加 x=0 的垂直线
plt.grid()
plt.legend()
plt.show()
```
