```python
# 导入库并绘制
from matplotlib_venn import venn2, venn3
import matplotlib.pyplot as plt
plt.rcParams["font.family"]=['SimHei'] # 用来设定字体样式
plt.rcParams['font.sans-serif']=['SimHei'] # 用来设定无衬线字体样式
plt.rcParams['axes.unicode_minus']=False # 用来正常显示负号
# 示例：2个集合的韦恩图
plt.figure(figsize=(6, 6))
venn2(subsets=(12, 15, 10), set_labels=('有纹身', '坏人'))
plt.title("坏人和有纹身的人的交集")
plt.show()

```

$$ P(有纹身|坏人) = \frac{P(有纹身 \cap 坏人)}{P(坏人)} $$
$$ P(坏人|有纹身) = \frac{P(有纹身 \cap 坏人)}{P(有纹身)} $$
$$ P(坏人|有纹身) \times P(有纹身) = P(有纹身|坏人) \times P(坏人) $$
$$ P(坏人|有纹身)  = \frac{P(有纹身|坏人) \times P(坏人)}{P(有纹身)} $$

```python
#假设总样本有 1000 个人.  坏人的概率 15/1000 = 0.015   , 有纹身的概率是 12 /1000 = 0.015  
# 坏人的概率
15/1000
```

```python
#坏人的概率 0.015  是坏人的先验概率
# p(有纹身 | 坏人) / p (有纹身)  如果大于1 , 则提高判断.  
# 意思是 在 坏人中 有纹身的概率 大于 普通人中的有纹身的概率.   就可以增加 对先验概率的判断. 说明纹身可以作为feature
```
