```python
# 三元一次方程
import numpy as np

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0, 0, 0, 1]]).T
syn0 = 2 * np.random.random((2, 1)) - 1
# 初始化偏置，范围在 [-1, 1] 之间
b = 2 * np.random.random((1, 1)) - 1

lr = 0.001
for j in range(10000):
    l0 = X
    l1 = np.dot(l0, syn0) + b  # 4*3 3*1 =  4*1
    l1_error = y - l1
    if j % 100 == 0:
        print(f"Iteration {j}, Error: {np.mean(l1_error**2):.10f}")
    syn0 += lr * np.dot(l0.T, l1_error)  #3*4 * 4*1  = 3*1
    b += lr * np.sum(l1_error, axis=0, keepdims=True)  # 1x1 + 4x1 = 1x1

print(syn0)
print(np.dot(X, syn0) + b)
print(b)
print(f"Y = {syn0[0,0]: .5f} * X1 + {syn0[1,0]: .5f} * X2 ")
```

```python
import numpy as np

# 与非门的输入数据
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# 与非门的目标输出
y = np.array([0, 0, 0, 1])

# 初始化权重和偏置
syn0 = 2 * np.random.random(2) - 1  # 2个输入对应的2个权重
b = 2 * np.random.random(1) - 1      # 1个偏置

# 学习率
lr = 0.1

# 感知机训练
for epoch in range(10000):
    for i in range(len(X)):
        # 计算加权和并加上偏置
        l1 = np.dot(X[i], syn0) + b
        
        # 感知机的激活函数：阶跃函数
        output = 1 if l1 >= 0 else 0
        
        # 计算误差
        error = y[i] - output
        
        # 更新权重和偏置
        syn0 += lr * error * X[i]
        b += lr * error

    # 每隔1000轮打印一次误差
    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Weights: {syn0}, Bias: {b}")

# 输出训练后的权重、偏置和模型对每个输入的预测
print("Trained Weights:", syn0)
print("Trained Bias:", b)
for i in range(len(X)):
    l1 = np.dot(X[i], syn0) + b
    output = 1 if l1 >= 0 else 0
    print(f"Input: {X[i]}, Predicted Output: {output}, Actual Output: {y[i]}")
```
