```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1.0, 5.0, 21)
y = [4, 4, 5, 6, 8, 12, 18, 22, 35, 40, 45, \
     33, 28, 30, 32, 36, 38, 41, 35, 31, 20]
y = [ yy/100 for yy in y]

# 定义曲线的 x 范围
x_curve = np.linspace(1.0, 5.0, 100)
# 定义曲线的 y 值
y_curve = 1 / (1 + np.exp(-(0.828 * x_curve - 4.006)))


# 绘制散点图
plt.scatter(x, y, label='Data Points')
# 绘制曲线
plt.plot(x_curve, y_curve, label='Curve: y = 1 / (1 + exp(-(0.828 * x - 4.006)))', color='red')
plt.scatter(x, y)
plt.xlabel('Quality')
plt.ylabel('rate')
plt.grid()
plt.show()
```
